 **1. What is the difference between a function and a method in Python?**
   - 1. Function
        - Definition: A function is a block of reusable code that is defined independently and can be called by its name.
        - Scope: Functions are not tied to any specific object or class.
        - Syntax: Defined using the def keyword.
   - 2. Method
      - Definition: A method is a function that is associated with an object and is called on that object.
      - Scope: Methods are defined within a class and operate on the instance of that class (or the class itself, in the case of class methods).
      - Syntax: Defined like a function but always takes self (or cls for class methods) as its first parameter.

 **2. Explain the concept of function arguments and parameters in Python**
 1. Parameters
     - Definition: Parameters are variables defined in the function's declaration (or definition). They act as placeholders for the values that will be passed to the function.
     - Example:
         - def greet(name):  # 'name' is a parameter
            -  print(f"Hello, {name}!")
2. Arguments
    - Definition: Arguments are the actual values you pass to the function when calling it. These values replace the parameters during the function's execution.
    - Example:
         - greet("Alice")  # "Alice" is the argument

3. Types of Arguments in Python:
   - positional arguments
   - keyword argument
   - default argument
   - variable-length argument
   - kwargs

In Python, functions can be defined and called in various ways depending on the use case.

1. Basic Function Definition and Call

   - Definition: Use the def keyword followed by the function name and parentheses.
   - Call: Use the function name followed by parentheses.

   - e.g.
     - def greet():

**4. What is the purpose of the `return` statement in a Python function?**

The return statement in a Python function serves the following purposes:

  - End Function Execution: It terminates the execution of the function as soon as it is encountered.
  - Return a Value: It sends a value (or multiple values) back to the caller of the function. This allows the function to produce a result that can be used elsewhere in the program.
  - Optional Usage: If no value is specified after return, the function returns None by default

**5. What are iterators in Python and how do they differ from iterables?**

Iterators:

An iterator is an object that represents a stream of data, allowing you to traverse through it one element at a time.
It implements two methods:

__iter__() (returns the iterator object itself).

__next__() (returns the next element in the sequence or raises StopIteration when no more elements are available).

Iterables:

An iterable is any object that can be looped over (e.g., using a for loop).

It implements the __iter__() method, which returns an iterator.

Examples of iterables: lists, tuples, strings, dictionaries, sets, etc.


**6. Explain the concept of generators in Python and how they are defined**

Generators in Python are a special type of function that return an iterator, which can be iterated over one value at a time. They are particularly useful when dealing with large datasets or streams of data, as they allow for lazy evaluation, meaning values are generated on-the-fly and not stored in memory all at once.

Creating Generators

To create a generator in Python, you define a function using the def keyword, but instead of using return to return a value, you use the yield keyword. This makes the function a generator function. When called, it returns a generator object that can be iterated over to produce the values.


**7. What are the advantages of using generators over regular functions?**

**Advantages of Generators**

  - Memory Efficiency: Generators generate values on-the-fly, reducing memory usage compared to storing the entire sequence in memory.

  - Lazy Evaluation: Values are computed only when needed, which can lead to performance improvements, especially with large datasets.

  - Maintain State: Generators automatically maintain their state between yield statements, making them suitable for iterative processes.

  - Convenience: Generators simplify code for iterating over large datasets or streams of data without needing to manage state explicitly.

  - Improved Performance: Because they avoid loading everything into memory up front, generators often run faster and scale better than regular functions with lists or other collections.

  - Cleaner Code: You can write more elegant and readable code for iteration-heavy tasks. Think: replacing nested loops with generator expressions or chaining together multiple generators.

  - Composable Pipelines: Generators can be chained together to form data-processing pipelines, each performing a transformationâ€”sort of like Unix pipes but in Python.

** 8. What is a lambda function in Python and when is it typically used?**

A lambda function in Python is a small, anonymous function defined using the lambda keyword. It can take any number of arguments but is limited to a single expression. The result of the expression is automatically returned.

Syntax:
Copy code
lambda arguments: expression

Typical Use Cases:

Short-lived Functions: Lambda functions are often used when a simple function is required temporarily, especially as an argument to higher-order functions.

With Built-in Functions:

filter(): To filter elements based on a condition
    
    numbers = [1, 2, 3, 4, 5]
    even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
    print(even_numbers)  # Output: [2, 4]


map(): To apply a function to all elements in a sequence.
    
    - numbers = [1, 2, 3, 4, 5]
    squares = list(map(lambda x: x ** 2, numbers))
    print(squares)  # Output: [1, 4, 9, 16, 25]


reduce() (from functools): To perform cumulative operations.


    from functools import reduce
      numbers = [1, 2, 3, 4]
      product = reduce(lambda x, y: x * y, numbers)
      print(product)  # Output: 24

Inline Functionality: When defining a full function using def feels unnecessary, such as in sorting or key-based operations.


    pairs = [(1, 'one'), (2, 'two'), (3, 'three')]
    pairs.sort(key=lambda x: x[1])
    print(pairs)  # Output: [(1, 'one'), (3, 'three'), (2, 'two')]


**9. Explain the purpose and usage of the `map()` function in Python.**

The map() function in Python is a powerful tool for applying a function to every item in an iterable, such as a list or tuple. It returns a map object, which is an iterator, allowing for efficient data transformation without creating intermediate lists.

Syntax

   map(function, iterable, ...)
   
   function: The function to apply to each element of the iterable.

iterable: The iterable(s) whose elements will be processed. Multiple iterables can be passed if the function accepts multiple arguments.

**10.What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?**

**1.map() Function**

The map() function applies a given function to each item in an iterable (like a list or tuple) and returns a map object (an iterator). It is commonly used for transforming data.

Syntax: map(function, iterable)

**2.filter() Function**

The filter() function filters elements from an iterable based on a function that returns True or False. It returns an iterator containing only the elements that satisfy the condition.

Syntax: filter(function, iterable)

**3.reduce() Function**

The reduce() function applies a function cumulatively to the items of an iterable, reducing it to a single value. Unlike map() and filter(), it is not built-in and must be imported from the functools module.

Syntax: reduce(function, iterable[, initial])

```python
# 4. Write a Python function that checks if a given number is prime or not from 1 to 200

def is_prime(number):
    if number <= 1:
        return False
    for i in range(2, int(number**0.5) + 1):
        if number % i == 0:
            return False
    return True

for num in range(1, 201):
   if is_prime(num):
    print(f"{num} is a prime number.")

```

```python
#6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
def powers_of_two(max_exponent):
    for exponent in range(max_exponent + 1):
        yield 2 ** exponent

for power in powers_of_two(7):
   print(power)

```

```python
#10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

def remove_vowels(input_string):
    vowels = "aeiouAEIOU"
    result = filter(lambda char: char not in vowels, input_string)
    return ''.join(result)
input_string="Hi my name is gargi how are YOU"
output_string= remove_vowels(input_string)
print("String without vowels:", output_string)
```
